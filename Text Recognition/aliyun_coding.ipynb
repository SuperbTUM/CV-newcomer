```python
import csv
with open("result.txt", "r") as f:
    lines = f.readlines()
    line_index = 11
    with open("detection_labels.csv", "w", encoding="utf-8", newline="") as g:
        writer = csv.writer(g)
        writer.writerow(["file_name", "file_code"])
        cur_row = []
        while line_index < len(lines):
            cur = lines[line_index]

            if "aliyun_yolov4" in cur:
                image_info = cur.split()
                image_index = image_info[0].split("/")[-1]
                cur_row = [image_index]
            elif cur[0] in "0123456789":
                cur_row.append(cur[0])
            else:
                if cur_row:
                    cur_res = [cur_row[0][:-1]] + ["".join(cur_row[1:])]
                    writer.writerow(cur_res)
                    cur_row = []
            line_index += 1
    g.close()
f.close()
```
